##**KAGGLE**

```python
import pandas as pd

data_kaggle = pd.read_excel('diabetes_kaggle.xlsx')
data_kaggle
```

```python
# Menghapus kolom yang tak diperlukan
kaggle_baru = data_kaggle.drop(columns=['hypertension','heart_disease','smoking_history','bmi'])
```

```python
# Mencari Missing Values dan Duplicates Data
kaggle_baru.info()
print('Total Missing Values: ', kaggle_baru.isna().sum().sum())
print('Total Duplicates Data: ', kaggle_baru.duplicated().sum())
```

```python
# Pisahkan kolom numerik dan kategori
numerik = kaggle_baru.select_dtypes(include=['float64', 'int64'])
kategori = kaggle_baru.select_dtypes(include=['object', 'category'])
```

```python
# Perbandingan jumlah data pada tipe data kategorikal
for col in kategori.columns:
    print(f"Distribusi pada kolom {col}:")
    print(kaggle_baru[col].value_counts())
    print("\n")
```

###Duplikat

```python
# Menghapus duplikat
kaggle_baru1 = kaggle_baru.drop_duplicates()

print("Data setelah menghapus duplikat:", kaggle_baru1.duplicated().sum())
print(kaggle_baru1)
```

###Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding
label_encoder = LabelEncoder()
kaggle_baru1['gender'] = label_encoder.fit_transform(kaggle_baru1['gender'])

print(kaggle_baru1)
```

###Outlier

```python
from sklearn.ensemble import IsolationForest

# Inisialisasi model Isolation Forest
model = IsolationForest(contamination=0.05)  # Menentukan persentase outliers yang diharapkan
outliers = model.fit_predict(kaggle_baru1)

# Mengubah hasil prediksi (-1 untuk outlier, 1 untuk normal)
kaggle_no_outliers = kaggle_baru1[outliers == 1]

# Menampilkan hasil tanpa outliers
print(kaggle_no_outliers)
```

```python
kaggle_no_outliers['diabetes'].value_counts()
```

```python
Xkaggle = kaggle_no_outliers.drop(columns=['diabetes'])
ykaggle = kaggle_no_outliers['diabetes']
```

```python
from sklearn.preprocessing import LabelEncoder

# Mengonversi semua kolom kategori menjadi angka
label_encoders = {}
for col in Xkaggle.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    Xkaggle[col] = le.fit_transform(Xkaggle[col])  # Mengubah kategori ke angka
    label_encoders[col] = le  # Menyimpan encoder jika perlu digunakan kembali
```

###SMOTE

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

print("Sebelum SMOTE:", Counter(ykaggle))

smote = SMOTE(sampling_strategy='auto', random_state=7, k_neighbors=5)
X_resampled, y_resampled = smote.fit_resample(Xkaggle, ykaggle)

print("Setelah SMOTE:", Counter(y_resampled))
```

##**PUSKESMAS**

```python
data_puskesmas = pd.read_excel('diabetes_puskesmas.xlsx')
data_puskesmas
```

```python
puskesmas_baru = data_puskesmas.drop(columns=['No','Nama','HCT (Hematokrit)','WBC (White Blood Cells)', 'RBC (Red Blood Cells)', 'PLT (Platelet)'])
```

```python
puskesmas_baru = puskesmas_baru.rename(columns={
    'Jenis Kelamin': 'gender',
    'Umur': 'age',
    'HbA1c': 'HbA1c_level',
    'Gula Darah': 'blood_glucose_level',
    'Label': 'diabetes'
})

puskesmas_baru
```

###Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding
label_encoder = LabelEncoder()
puskesmas_baru['gender'] = label_encoder.fit_transform(puskesmas_baru['gender'])
puskesmas_baru['diabetes'] = label_encoder.fit_transform(puskesmas_baru['diabetes'])

print(puskesmas_baru)
```

```python
Xpuskesmas = puskesmas_baru.drop(columns=['diabetes'])
ypuskesmas = puskesmas_baru['diabetes']
```

```python
from sklearn.preprocessing import LabelEncoder

# Mengonversi semua kolom kategori menjadi angka
label_encoders = {}
for col in Xpuskesmas.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    Xpuskesmas[col] = le.fit_transform(Xpuskesmas[col])  # Mengubah kategori ke angka
    label_encoders[col] = le  # Menyimpan encoder jika perlu digunakan kembali
```

##**PEMODELAN**

###AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score

# Inisialisasi model AdaBoost dengan parameter default
adaboost = AdaBoostClassifier()

# Melatih model dengan data Kaggle
adaboost.fit(Xkaggle, ykaggle)

# Prediksi pada data Kaggle
y_pred_kaggle = adaboost.predict(Xkaggle)

# Menghitung akurasi
accuracy_kaggle = accuracy_score(ykaggle, y_pred_kaggle)
print("Akurasi model AdaBoost pada data Kaggle:", accuracy_kaggle)
```

###AdaBoost+SMOTE

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score

# Inisialisasi model AdaBoost dengan parameter default
adaboost = AdaBoostClassifier()

# Melatih model dengan data Kaggle yang telah di-balancing menggunakan SMOTE
adaboost.fit(X_resampled, y_resampled)

# Prediksi pada data Kaggle
y_pred_kaggle = adaboost.predict(Xkaggle)

# Menghitung akurasi pada data Kaggle
accuracy_kaggle = accuracy_score(ykaggle, y_pred_kaggle)
print("Akurasi model AdaBoost + SMOTE pada data Kaggle:", accuracy_kaggle)
```

###RandomSearchCV+AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score

# Inisialisasi model AdaBoost
adaboost = AdaBoostClassifier()

# Definisi ruang hyperparameter
param_distributions = {
    'n_estimators': range(10, 101, 10),
    'learning_rate': [0.01, 0.1, 0.5, 1]
}

# Inisialisasi RandomizedSearchCV dengan data Kaggle sebelum SMOTE
random_search = RandomizedSearchCV(
    estimator=adaboost,
    param_distributions=param_distributions,
    n_iter=20,  # Jumlah kombinasi yang diuji
    scoring='accuracy',  # Gunakan metrik evaluasi yang sesuai
    cv=5,  # 5-fold cross-validation
    random_state=7,
    n_jobs=-1  # Gunakan semua core untuk komputasi
)

# Melatih model dengan data Kaggle sebelum SMOTE
random_search.fit(Xkaggle, ykaggle)

# Menampilkan hasil terbaik
print("Best Parameters:", random_search.best_params_)
print("Best Cross-Validation Score:", random_search.best_score_)

# Gunakan model terbaik untuk prediksi pada data Kaggle
best_adaboost = random_search.best_estimator_
y_pred_kaggle = best_adaboost.predict(Xkaggle)

# Evaluasi akurasi pada data Kaggle
accuracy_kaggle = accuracy_score(ykaggle, y_pred_kaggle)
print("Akurasi model terbaik AdaBoost pada data Kaggle:", accuracy_kaggle)
```

###RandomSearchCV+AdaBoost+SMOTE

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score
import pandas as pd

# Inisialisasi model AdaBoost
adaboost = AdaBoostClassifier()

# Definisi ruang hyperparameter
param_distributions = {
    'n_estimators': range(10, 101, 10),
    'learning_rate': [0.01, 0.1, 0.5, 1]
}

# Inisialisasi RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=adaboost,
    param_distributions=param_distributions,
    n_iter=20,  # Jumlah kombinasi yang diuji
    scoring='accuracy',  # Gunakan metrik evaluasi yang sesuai
    cv=5,  # 5-fold cross-validation
    random_state=7,
    n_jobs=-1  # Gunakan semua core untuk komputasi
)

# Jalankan pencarian hyperparameter menggunakan data training (Kaggle setelah SMOTE)
random_search.fit(X_resampled, y_resampled)

# Menampilkan semua hasil pencarian hyperparameter
results = pd.DataFrame(random_search.cv_results_)
results = results[['param_n_estimators', 'param_learning_rate', 'mean_test_score']]
print("\nHasil Pencarian Hyperparameter:")
print(results.sort_values(by="mean_test_score", ascending=False))  # Urutkan berdasarkan skor tertinggi

# Gunakan model terbaik untuk prediksi pada data Kaggle
best_adaboost1 = random_search.best_estimator_
y_pred_kaggle = best_adaboost1.predict(Xkaggle)

# Evaluasi akurasi pada data Kaggle
accuracy_kaggle = accuracy_score(ykaggle, y_pred_kaggle)
print("\nAkurasi model terbaik AdaBoost pada data Kaggle:", accuracy_kaggle)
```

```python
import joblib
from sklearn.metrics import accuracy_score, classification_report

# Load model AdaBoost yang telah disimpan
model = joblib.load("adaboost_best_model.pkl")

# Lakukan prediksi pada data Puskesmas
y_pred = model.predict(Xpuskesmas)

# Evaluasi hasil prediksi
accuracy = accuracy_score(ypuskesmas, y_pred)
report = classification_report(ypuskesmas, y_pred)

print(f"Akurasi pada data Puskesmas: {accuracy * 100:.2f}%")
print("Classification Report:\n", report)
```
